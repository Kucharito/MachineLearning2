# K-Means zhlukovanie zákazníkov

```python
from sklearn.preprocessing import StandardScaler
features = ['Income', 'Customer_Age', 'Children', 'Recency',
            'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
            'NumWebVisitsMonth', 'Total_Spent']

X = df[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df['Cluster'] = kmeans.fit_predict(X_scaled)
print(df['Cluster'].value_counts())
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)

df['PCA1'] = pca_result[:, 0]
df['PCA2'] = pca_result[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df, palette='Set2', alpha=0.7)
plt.title('Customer Segments based on PCA')
plt.show()
```

```python
cluster_summary = df.groupby('Cluster')[features].mean().round(2)
print(cluster_summary)
```

### Cluster 1 má najvyšší priemerný Total_Spent (okolo 1380) — 🟢 bohatí a aktívni zákazníci.
### Cluster 2 má nízke výdavky (~143) — 🟠 nízke príjmy, menej aktívni zákazníci.
### Cluster 0 je ešte nižší (~100) — 🔵 mladší zákazníci s nízkym príjmom, zriedkavé nákupy.
### Cluster 3 má strednú úroveň výdavkov (~800) — 🟣 rodiny s priemerným príjmom, stabilné nákupy.

# Hierarchické zhlukovanie a vizualizácia, dendrogram

```python
from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=4, linkage = 'ward')
df['HC_Cluster'] = hc.fit_predict(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='HC_Cluster', data=df, palette='Set1', alpha=0.7)
plt.title('Hierarchical Clustering Customer Segments based on PCA')
plt.show()
```

```python
hc_summary = df.groupby('HC_Cluster')[[
    'Income', 'Total_Spent', 'Customer_Age', 'Children',
    'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases'
]].mean().round(2)

print(hc_summary)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

plt.figure(figsize=(8,5))
Z = linkage(X_scaled, method='ward')
dendrogram(Z, truncate_mode='level', p=5)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

Hierarchické zhlukovanie pomocou Wardovej metódy prinieslo výsledky, ktoré sa veľmi podobali segmentácii K-Means.
Objavili sa tie isté štyri hlavné skupiny zákazníkov, čo potvrdzuje, že oba prístupy identifikujú podobnú štruktúru v údajoch.
V oboch metódach môžeme jasne vidieť rozdiel medzi zákazníkmi s vysokými a nízkymi výdavkami, ako aj výrazné rozdiely vo veku, príjme a nákupnom správaní.
Táto konzistentnosť naznačuje, že objavené zhluky sú stabilné a významné pre pochopenie profilov zákazníkov.


# DBSCAN zhlukovanie a detekcia odľahlých hodnôt

```python
from sklearn.cluster import DBSCAN

dbscan = DBSCAN(eps=1.2, min_samples=5)
df['DBSCAN_Cluster'] = dbscan.fit_predict(X_scaled)

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='DBSCAN_Cluster', data=df, palette='Set3', alpha=0.7)
plt.title('DBSCAN Clustering Customer Segments based on PCA')
plt.show()
```

Metóda DBSCAN rozdelila zákazníkov podľa hustoty dát do viacerých menších skupín.
Väčšina bodov spadá do niekoľkých hustých oblastí, ktoré zodpovedajú hlavným typom zákazníkov,
zatiaľ čo niekoľko bodov bolo identifikovaných ako odľahlé hodnoty (-1) – ide prevažne o zákazníkov s mimoriadne vysokým príjmom alebo výdavkami.
Na rozdiel od metódy K-Means DBSCAN netvorí rovnako veľké klastre,
ale umožňuje lepšie odhaliť anomálie a lokálne odchýlky v správaní zákazníkov
