Pri clusteringu (K-Means, Hierarchical, DBSCAN) chceme zahrnúť len také číselné atribúty, ktoré vystihujú správanie alebo charakter zákazníka.
Nie všetky atribúty z pôvodného datasetu sú na to vhodné — niektoré sú:

identifikačné (ID) – nepatria tam,

konštantné (Z_CostContact, Z_Revenue) – neovplyvňujú správanie,

výsledky kampaní (AcceptedCmpX, Response) – používajú sa len neskôr na interpretáciu, nie na tvorbu klastrov.

```python
from sklearn.preprocessing import OneHotEncoder

categorical_cols = ['Education', 'Marital_Status', 'Living_With']
numeric_cols = [
    'Income', 'Customer_Age', 'Recency', 'Total_Spent',
    'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
    'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases',
    'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
    'NumWebVisitsMonth', 'Children', 'FamilySize'
]

encoder = OneHotEncoder(sparse_output=False, drop='first')
encoded = encoder.fit_transform(df[categorical_cols])
encoded_cols = encoder.get_feature_names_out(categorical_cols)
encoded_df = pd.DataFrame(encoded, columns=encoded_cols)

df_final = pd.concat([df[numeric_cols], encoded_df, df['Response']], axis=1)
print(df_final.head())
```

```python
from sklearn.preprocessing import StandardScaler

df_final = df_final.dropna()
print("Missing values in df_final:", df_final.isnull().sum().sum())

# škálovanie číselných hodnôt (všetko okrem 'Response')
scaler = StandardScaler()
scaled = scaler.fit_transform(df_final.drop('Response', axis=1))

# vytvorenie DataFrame po škálovaní
X = pd.DataFrame(scaled, columns=df_final.drop('Response', axis=1).columns)
print(X.head())
print("Missing values in X:", X.isnull().sum().sum())


```

# K-Means zhlukovanie zákazníkov

```python
from sklearn.preprocessing import StandardScaler
features = ['Income', 'Customer_Age', 'Recency', 'Total_Spent',
    'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
    'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases',
    'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
    'NumWebVisitsMonth', 'Children', 'FamilySize']

X = df_final[features]
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)
```

```python
from sklearn.cluster import KMeans
import matplotlib.pyplot as plt

inertia = []
K = range(2, 11)

for k in K:
    kmeans = KMeans(n_clusters=k, random_state=42)
    kmeans.fit(X_scaled)
    inertia.append(kmeans.inertia_)

plt.figure(figsize=(8, 5))
plt.plot(K, inertia, 'bo-')
plt.xlabel('Number of clusters (k)')
plt.ylabel('Inertia')
plt.title('Elbow Method For Optimal k')
plt.show()
```

```python
kmeans = KMeans(n_clusters=4, random_state=42)
df_final['Cluster'] = kmeans.fit_predict(X_scaled)
print(df_final['Cluster'].value_counts())
```

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2)
pca_result = pca.fit_transform(X_scaled)

df_final['PCA1'] = pca_result[:, 0]
df_final['PCA2'] = pca_result[:, 1]

plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='Cluster', data=df_final, palette='Set2', alpha=0.7)
plt.title('Customer Segments based on PCA')
plt.show()
```

```python
cluster_summary = df_final.groupby('Cluster')[features].mean().round(2)
print(cluster_summary)
```

# Hierarchické zhlukovanie a vizualizácia, dendrogram

```python
from sklearn.preprocessing import StandardScaler

features = ['Income', 'Customer_Age', 'Recency', 'Total_Spent',
    'MntWines', 'MntFruits', 'MntMeatProducts', 'MntFishProducts',
    'MntSweetProducts', 'MntGoldProds', 'NumDealsPurchases',
    'NumWebPurchases', 'NumCatalogPurchases', 'NumStorePurchases',
    'NumWebVisitsMonth', 'Children', 'FamilySize']

scaler = StandardScaler()
X_scaled = scaler.fit_transform(df_final[features])

from sklearn.cluster import AgglomerativeClustering

hc = AgglomerativeClustering(n_clusters=4, linkage='ward')
df_final['HC_Cluster'] = hc.fit_predict(X_scaled)

# a vizualizáciu
plt.figure(figsize=(10, 6))
sns.scatterplot(x='PCA1', y='PCA2', hue='HC_Cluster', data=df_final, palette='Set1', alpha=0.7)
plt.title('Hierarchical Clustering Customer Segments based on PCA')
plt.show()
```

```python
print(X_scaled.shape)
print(df_final[features].shape)
```

```python
from scipy.cluster.hierarchy import linkage, dendrogram

plt.figure(figsize=(8,5))
Z = linkage(X_scaled, method='ward')
dendrogram(Z, truncate_mode='level', p=5)
plt.title('Dendrogram for Hierarchical Clustering')
plt.xlabel('Samples')
plt.ylabel('Distance')
plt.show()
```

```python
df_final.groupby('HC_Cluster')[features].mean().round(2)
```

##  DBSCAN – Hĺbková analýza zhlukov

Model **DBSCAN (Density-Based Spatial Clustering of Applications with Noise)** dokáže rozpoznať aj nepravidelné skupiny zákazníkov a odhaliť **outliery** (označené ako `-1`).  
Na základe parametrov `eps=1.2` a `min_samples=5` vzniklo **8 zhlukov** + **1 skupina outlierov**.

---

###  Silhouette Score
Výsledné skóre: **Silhouette = 0.146**, čo predstavuje **slabšiu kvalitu segmentácie**.  
Zhluky sa **čiastočne prekrývajú**, a teda DBSCAN **neodhalil veľmi jasnú štruktúru** v dátach.  
Tento výsledok je typický pre marketingové dáta s premenlivou hustotou a spojitými číselnými hodnotami.
